## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Data Wrangling

# Remove Duplicate Rows

for name, df in datasets.items():
    df.drop_duplicates(inplace=True)
    print(f"Duplicates removed from {name} dataset")


# Standardize Column Names

for name, df in datasets.items():
    df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
    print(f"Column names standardized for {name} dataset")


# Handle Missing Values

for name, df in datasets.items():
    
    # Numerical Columns
    num_cols = df.select_dtypes(include=np.number).columns
    
    # Categorical Columns
    cat_cols = df.select_dtypes(include='object').columns
    
    # Fill Numerical Missing Values
    for col in num_cols:
        df[col].fillna(df[col].median(), inplace=True)
    
    # Fill Categorical Missing Values
    for col in cat_cols:
        df[col].fillna(df[col].mode()[0], inplace=True)
    
    print(f"Missing values handled for {name} dataset")


# Standardize Text Columns

for name, df in datasets.items():
    
    cat_cols = df.select_dtypes(include='object').columns
    
    for col in cat_cols:
        df[col] = df[col].astype(str).str.strip().str.title()
    
    print(f"Text values standardized for {name} dataset")


# Handle Incorrect Ratings

if 'rating' in transaction_df.columns:
    
    transaction_df = transaction_df[
        (transaction_df['rating'] >= 1) &
        (transaction_df['rating'] <= 5)
    ]
    
    print("Invalid ratings removed")


# Reset Index

for name, df in datasets.items():
    df.reset_index(drop=True, inplace=True)

print("Data Wrangling Completed Successfully")
````

### Date standardization

````python
# Standardize Date Format

transaction_df['visit_date'] = pd.to_datetime(
    transaction_df['visityear'].astype(str) + '-' +
    transaction_df['visitmonth'].astype(str),
    format='%Y-%m'
)

transaction_df['visit_date'] = transaction_df['visit_date'].dt.strftime('%Y-%m')

transaction_df[['visityear', 'visitmonth', 'visit_date']].head()
````

### Outlier handling

````python
# Outlier Detection using Boxplot

plt.figure(figsize=(8,5))

sns.boxplot(x=transaction_df['rating'])

plt.title("Rating Outliers")

plt.show()
````

````python
# Handle Outliers using IQR Method

Q1 = transaction_df['rating'].quantile(0.25)
Q3 = transaction_df['rating'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

transaction_df = transaction_df[
    (transaction_df['rating'] >= lower_bound) &
    (transaction_df['rating'] <= upper_bound)
]

print("Outliers Removed Successfully")
````

### Merging Dataset


````python
# Merge Datasets

merged_df = transaction_df.merge(
    user_df,
    on='userid',
    how='left'
)

merged_df = merged_df.merge(
    city_df,
    on='cityid',
    how='left',
    suffixes=('', '_city')
)

merged_df = merged_df.merge(
    country_df,
    on='countryid',
    how='left',
    suffixes=('', '_country')
)

merged_df = merged_df.merge(
    region_df,
    on='regionid',
    how='left',
    suffixes=('', '_region')
)

merged_df = merged_df.merge(
    continent_df,
    on='continentid',
    how='left',
    suffixes=('', '_continent')
)

merged_df = merged_df.merge(
    item_df,
    on='attractionid',
    how='left'
)

merged_df = merged_df.merge(
    type_df,
    on='attractiontypeid',
    how='left'
)

print("Datasets Merged Successfully")

merged_df.head()
````

### Feature engineering

````python
# Feature Engineering

# Average Rating Per User

user_avg_rating = merged_df.groupby('userid')['rating'].mean().reset_index()

user_avg_rating.columns = ['userid', 'avg_user_rating']

merged_df = merged_df.merge(
    user_avg_rating,
    on='userid',
    how='left'
)

print("Feature Engineering Completed")

merged_df.head()
````

### Feature Encoding

````python
# Encode Categorical Variables

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

categorical_columns = [
    'visitmode',
    'continent',
    'country',
    'region',
    'cityname',
    'attractiontype'
]

for col in categorical_columns:
    
    merged_df[col] = le.fit_transform(
        merged_df[col].astype(str)
    )

print("Categorical Encoding Completed")
````

### Scaling/Normalization

````python
# Feature Scaling / Normalization

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

numerical_columns = [
    'rating',
    'avg_user_rating'
]

merged_df[numerical_columns] = scaler.fit_transform(
    merged_df[numerical_columns]
)

print("Feature Scaling Completed")

merged_df.head()
````

## ***5. Hypothesis Testing***

### Based on your chart experiments, define three hypothetical statements from the dataset. In the next three questions, perform hypothesis testing to obtain final conclusion about the statements through your code and statistical testing.

Based on chart experiments, three hypothetical statements were defined to analyze the relationship between tourist behavior, attraction characteristics, and user demographics. Statistical hypothesis testing was performed using ANOVA and Chi-Square tests to determine whether significant relationships exist among the selected variables. These tests help validate the patterns observed during exploratory data analysis and support data-driven conclusions for tourism prediction and recommendation systems.

### Hypothetical Statement - 1

There is a significant relationship between Visit Mode and Attraction Ratings.

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

H₀ (Null): There is no significant relationship between Visit Mode and Attraction Ratings.

H₁ (Alternate): There is a significant relationship between Visit Mode and Attraction Ratings.

#### 2. Perform an appropriate statistical test.

````python
# Hypothesis Testing - 1
# ANOVA Test

from scipy.stats import f_oneway

visitmode_groups = [
    group['rating'].values
    for name, group in merged_df.groupby('visitmode')
]

f_stat, p_value = f_oneway(*visitmode_groups)

print("F-Statistic :", f_stat)
print("P-Value :", p_value)

if p_value < 0.05:
    print("Reject Null Hypothesis")
else:
    print("Fail to Reject Null Hypothesis")
````

##### Which statistical test have you done to obtain P-Value?

One-Way ANOVA

##### Why did you choose the specific statistical test?

ANOVA is suitable because it compares the mean ratings across multiple Visit Mode categories and determines whether significant differences exist between them.

### Hypothetical Statement - 2

Different attraction types receive significantly different tourist ratings.

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

H₀: There is no significant difference in ratings among attraction types.

H₁: There is a significant difference in ratings among attraction types.

#### 2. Perform an appropriate statistical test.

````python
# Hypothesis Testing - 2
# ANOVA Test

attraction_groups = [
    group['rating'].values
    for name, group in merged_df.groupby('attractiontype')
]

f_stat, p_value = f_oneway(*attraction_groups)

print("F-Statistic :", f_stat)
print("P-Value :", p_value)

if p_value < 0.05:
    print("Reject Null Hypothesis")
else:
    print("Fail to Reject Null Hypothesis")
````

##### Which statistical test have you done to obtain P-Value?

One-Way ANOVA

##### Why did you choose the specific statistical test?

ANOVA is appropriate because it compares average ratings across multiple attraction type categories to identify significant differences.

### Hypothetical Statement - 3

User demographics significantly influence Visit Mode.

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

H₀: User demographics do not significantly influence Visit Mode.

H₁: User demographics significantly influence Visit Mode.

#### 2. Perform an appropriate statistical test.

````python
# Hypothesis Testing - 3
# Chi-Square Test

from scipy.stats import chi2_contingency

contingency_table = pd.crosstab(
    merged_df['continent'],
    merged_df['visitmode']
)

chi2, p_value, dof, expected = chi2_contingency(contingency_table)

print("Chi-Square Statistic :", chi2)
print("P-Value :", p_value)

if p_value < 0.05:
    print("Reject Null Hypothesis")
else:
    print("Fail to Reject Null Hypothesis")
````

##### Which statistical test have you done to obtain P-Value?

Chi-Square Test of Independence was performed.

##### Why did you choose the specific statistical test?

The Chi-Square test is suitable because both Continent and Visit Mode are categorical variables, and the test helps determine whether they are statistically associated.

The user-item matrix is the foundation of collaborative filtering recommendation systems. It represents user preferences by storing ratings given by users to attractions. This matrix helps identify similar users and recommend attractions based on shared interests and rating patterns.

#### Why is cosine similarity used?

Cosine similarity is used to measure the similarity between users based on their attraction ratings. Users with similar rating patterns are considered similar users, allowing the recommendation system to suggest attractions preferred by users with comparable interests and travel behavior.

````python
# Calculating User Similarity Matrix

from sklearn.metrics.pairwise import cosine_similarity

# Cosine Similarity

user_similarity = cosine_similarity(
    user_item_matrix
)

# Convert to DataFrame

user_similarity_df = pd.DataFrame(
    user_similarity,
    index=user_item_matrix.index,
    columns=user_item_matrix.index
)

print("User Similarity Matrix Created Successfully")

user_similarity_df.head()
````

#### Why is the recommendation function important?

The recommendation function identifies similar users based on rating behavior and suggests attractions that highly similar users preferred. This helps provide personalized attraction recommendations and improves user engagement and travel experience.

````python
# Recommendation Function

def recommend_attractions(user_id, top_n=5):
    
    # Similar Users
    
    similar_users = user_similarity_df[user_id].sort_values(
        ascending=False
    )[1:6]
    
    # Attractions Rated by Similar Users
    
    similar_user_ratings = user_item_matrix.loc[
        similar_users.index
    ]
    
    # Average Ratings
    
    recommendation_scores = similar_user_ratings.mean().sort_values(
        ascending=False
    )
    
    # Top Recommendations
    
    recommendations = recommendation_scores.head(top_n)
    
    return recommendations

# Example Recommendation

recommend_attractions(user_id=user_item_matrix.index[0])
````

#### Why is RMSE used for recommendation evaluation?

RMSE (Root Mean Squared Error) measures the difference between actual user ratings and predicted ratings. Lower RMSE values indicate better recommendation accuracy and improved prediction quality of the recommendation system.

````python
# Recommendation System Evaluation using RMSE

from sklearn.metrics import mean_squared_error

# Actual Ratings

actual_ratings = merged_df['rating']

# Predicted Ratings
# Using average user ratings from recommendation matrix

predicted_ratings = merged_df.groupby(
    'userid'
)['rating'].transform('mean')

# RMSE Calculation

rmse_recommendation = np.sqrt(
    mean_squared_error(
        actual_ratings,
        predicted_ratings
    )
)

print("Recommendation System RMSE :", rmse_recommendation)
````

````python
# Content-Based Filtering

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Keep Unique Attractions Only

content_df = merged_df[
    ['attraction', 'attractiontype', 'cityname', 'country']
].drop_duplicates()

# Fill Missing Values

content_df = content_df.fillna('unknown')

# Combine Features

content_df['content_features'] = (
    content_df['attractiontype'].astype(str) + " " +
    content_df['cityname'].astype(str) + " " +
    content_df['country'].astype(str)
)

# Reset Index

content_df = content_df.reset_index(drop=True)

# Vectorization

cv = CountVectorizer()

content_matrix = cv.fit_transform(
    content_df['content_features']
)

# Cosine Similarity

content_similarity = cosine_similarity(
    content_matrix
)

print("Optimized Content-Based Similarity Matrix Created Successfully")
````

#### Why is content-based recommendation important?

Content-based recommendation suggests attractions that are similar in type, city, and country to attractions previously preferred by the user. This improves personalization and helps users discover attractions matching their interests and travel preferences.

````python
# Content-Based Recommendation Function

def content_based_recommendation(attraction_name, top_n=5):
    
    # Find Attraction Index
    
    attraction_index = content_df[
        content_df['attraction'] == attraction_name
    ].index[0]
    
    # Similarity Scores
    
    similarity_scores = list(
        enumerate(content_similarity[attraction_index])
    )
    
    # Sort Similar Attractions
    
    sorted_scores = sorted(
        similarity_scores,
        key=lambda x: x[1],
        reverse=True
    )
    
    # Top Similar Attractions
    
    recommended_indices = [
        i[0]
        for i in sorted_scores[1:top_n+1]
    ]
    
    # Recommended Attractions
    
    recommendations = content_df.iloc[
        recommended_indices
    ]['attraction']
    
    return recommendations

# Example Recommendation

content_based_recommendation(
    content_df['attraction'].iloc[0]
)
````
